## PCM - Muestreo

Desarrolle un programa que para una señal de mensaje $m(t)$ consistente en una sinusoide o suma de sinusoides realice el proceso de muestreo PCM aplicando el teorema del muestreo $(f_s \geq 2f_M)$.

Ejercicio sacado de CyM Tech **[[YouTube](https://www.youtube.com/watch?v=bYdPrk6o2aM)]**

```python
import matplotlib.pyplot as plt
import numpy as np

#Definición de parámetros

t = np.linspace(0, 1, 1000)
x_t = 2*np.cos(2*np.pi*2*t)+3*np.sin(2*np.pi*6*t)


plt.subplot(4,1,1)
plt.plot(t, x_t)
plt.grid("True")
plt.title("Señal continua vs Señal Muestreada")
plt.show

F_s = 10
x = np.linspace(0,1,F_s)
d = 1+x-x
plt.subplot(4,1,2)
plt.stem(x, d)
plt.ylim(-5.5,5.5)
plt.grid("True")

T_s =1/F_s #Periodo de muestreo en segundos
nT = np.linspace(0, 1, F_s)
x_nT = 2*np.cos(2*np.pi*2*nT)+3*np.sin(2*np.pi*6*nT)

plt.subplot(4,1,3)
plt.stem(nT, x_nT)
plt.ylim(-5.5,5.5)
plt.grid("True")

N=F_s
n=np.linspace(0, N, N+1)
x_n = 2*np.cos(2*np.pi*(2/F_s)*n)+3*np.sin(2*np.pi*(6/F_s)*n)

plt.subplot(4,1,4)
plt.stem(n, x_n)
plt.ylim(-5.5,5.5)
plt.grid("True")
```
