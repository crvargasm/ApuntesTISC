## PCM - Muestreo

Desarrolle un programa que para una señal de mensaje $m(t)$ consistente en una sinusoide o suma de sinusoides realice el proceso de muestreo PCM aplicando el teorema del muestreo $(f_s \geq 2f_M)$.

$$x_t = 2\cos(2\pi 2 t) + 3\sin(2\pi 6t)$$

Ejercicio sacado de CyM Tech **[[YouTube](https://www.youtube.com/watch?v=bYdPrk6o2aM)]**

```python
import matplotlib.pyplot as plt
import numpy as np

#Señal de tiempo "Continuo"
t = np.linspace(0, 1, 1000)     #Tiempo desde 0 hasta 1 (sg) con 1000 steps entre ese intervalo
x_t = 2*np.cos(2*np.pi*2*t)+3*np.sin(2*np.pi*6*t)   #Abstracción de nuestra función de tiempo

plt.subplot(3,1,1)     
plt.plot(t, x_t)
plt.grid("True")
plt.title("Señal 'continua' en función del tiempo")
plt.show

```

```python
#Discretizacion de la señal análoga
F_s = 50    #frecuencia de muestreo [muestras por segundo]
T_s = 1/F_s  #Periodo de muestreo en segundos   [ciclos por segundo]
nT = np.linspace(0, 1, F_s)     #Instantes de muestreo
x_nT = 2*np.cos(2*np.pi*2*nT)+3*np.sin(2*np.pi*6*nT) #Función de tiempo discreto, cambiamos la variable de tiempo de 'continua' a discreta

plt.subplot(3,1,2)
plt.stem(nT, x_nT)
plt.ylim(-5.5,5.5)
plt.title("Señal muestrada en función del tiempo")
plt.grid("True")
```

```python
#Señal de tiempo discreto
N=F_s   #Número de muestras que vamos a tener de una señal en tiempo discreto
n=np.linspace(0, N, N)  #Conjunto de muestras [0,1,2,3,...,N]
x_n = 2*np.cos(2*np.pi*(2)*n)+3*np.sin(2*np.pi*(6)*n) #[ciclos por muestra]

plt.subplot(3,1,3)
plt.stem(n, x_n)
plt.ylim(-5.5,5.5)
plt.title("Señal muestrada en función de intervalos")
plt.grid("True")
```
